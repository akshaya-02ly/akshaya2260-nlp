LOAD WORD 2 VEC AND COMPLETE WORD SIMILARITY

```python
# Load Word2Vec and find word similarity

import gensim.downloader as api

# Load pre-trained Word2Vec model
model = api.load("word2vec-google-news-300")

# Word similarity
similarity = model.similarity("king", "queen")

print("Similarity between 'king' and 'queen':", similarity)
```

LOAD GLOVE AND COMPLETE WORD SIMILARITY

```python
# Load GloVe and find word similarity

import gensim.downloader as api

# Load pre-trained GloVe model
model = api.load("glove-wiki-gigaword-100")

# Word similarity
similarity = model.similarity("king", "queen")

print("Similarity between 'king' and 'queen':", similarity)
```

LOAD WORD 2 VEC AND FIND NEIGHBOUR WORDS

```python
# Load Word2Vec and find Neighbour Words

import gensim.downloader as api

# Load Word2Vec model
w2v_model = api.load("word2vec-google-news-300")

# Neighbour words for 'computer'
neighbors_w2v = w2v_model.most_similar("computer", topn=5)

print("Word2Vec Neighbours for 'computer':")
for word, score in neighbors_w2v:
    print(word, score)
```

LOAD GLOVE AND FIND NEIGHBOUR WORDS

```python
# Load GloVe and find Neighbour Words

import gensim.downloader as api

# Load GloVe model
glove_model = api.load("glove-wiki-gigaword-100")

# Neighbour words for 'computer'
neighbors_glove = glove_model.most_similar("computer", topn=5)

print("GloVe Neighbours for 'computer':")
for word, score in neighbors_glove:
    print(word, score)
```

LOAD WORD 2 VEC AND PERFORM WORD ANALOGY

```python
# Load Word2Vec and perform Word Analogy

import gensim.downloader as api

# Load Word2Vec model
w2v_model = api.load("word2vec-google-news-300")

# king - man + woman = ?
analogy_w2v = w2v_model.most_similar(positive=["king", "woman"], negative=["man"], topn=3)

print("Word2Vec Analogy (king - man + woman):")
for word, score in analogy_w2v:
    print(word, score)
```

LOAD GLOVE AND PERFORM WORD ANALOGY

```python
# Load GloVe and perform Word Analogy

import gensim.downloader as api

# Load GloVe model
glove_model = api.load("glove-wiki-gigaword-100")

# king - man + woman = ?
analogy_glove = glove_model.most_similar(positive=["king", "woman"], negative=["man"], topn=3)

print("GloVe Analogy (king - man + woman):")
for word, score in analogy_glove:
    print(word, score)
```
