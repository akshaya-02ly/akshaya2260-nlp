````python
data = {
    'text': [
        'I love this product! It is amazing and works perfectly.',
        'This is a terrible experience. I regret buying it.',
        'The service was excellent, very helpful and fast.',
        'Not satisfied with the quality, it broke after a week.',
        'Highly recommend! Best purchase ever.',
        'Could be better, some features are missing.',
        'Fantastic movie, a must-watch for everyone.',
        'Absolutely disgusted by the poor customer support.',
        'Pretty good for the price, no major complaints.',
        'Worst decision ever to buy this, complete waste of money.'
    ],
    'label': [
        'positive', 'negative', 'positive', 'negative', 'positive',
        'negative', 'positive', 'negative', 'positive', 'negative'
    ]
}
df = pd.DataFrame(data)

print("Sample DataFrame created successfully.")
print(df.head())
````

## Preprocess Text Data

### Subtask:
Clean the text data by converting it to lowercase, removing punctuation, numbers, and special characters, and then removing common English stopwords using NLTK. Display a sample of the processed text.


**Reasoning**:
Define a function to preprocess text by converting to lowercase, removing unwanted characters, tokenizing, and removing stopwords. Apply this function to the 'text' column of the DataFrame and display the original and processed text.



````python
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation, numbers, and special characters, keeping only alphabetic characters and spaces
    text = re.sub(r'[^a-z\s]', '', text)
    # Tokenize the text
    words = text.split()
    # Get English stopwords
    stop_words = set(stopwords.words('english'))
    # Remove stopwords
    words = [word for word in words if word not in stop_words]
    # Join the processed words back into a string
    text = ' '.join(words)
    return text

# Apply the preprocessing function to the 'text' column
df['processed_text'] = df['text'].apply(preprocess_text)

print("Text preprocessing complete. Displaying a sample of original and processed text:")
print(df[['text', 'processed_text']].head())
````

````python
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(df['processed_text'])

print("Shape of TF-IDF matrix (documents, vocabulary size):")
print(X_tfidf.shape)

vocabulary_size = len(vectorizer.vocabulary_)
print(f"Size of the generated vocabulary: {vocabulary_size}")
````

## Explain TF-IDF Weighting

### Subtask:
Explain the concept and intuition behind TF-IDF (Term Frequency-Inverse Document Frequency) weighting.

#### Explanation of TF-IDF Weighting

**TF-IDF** (Term Frequency-Inverse Document Frequency) is a numerical statistic that reflects how important a word is to a document in a collection or corpus. It is widely used as a weighting factor in information retrieval and text mining. The intuition behind TF-IDF is that words that appear frequently in a document but rarely in the entire corpus are likely to be very relevant to that specific document.

TF-IDF is composed of two main components:

1.  **Term Frequency (TF)**
    *   **Concept**: Term Frequency measures how frequently a term (word) appears in a document. The more often a word appears, the more relevant it is considered to be to that document.
    *   **Intuition**: If a word occurs many times in a document, it probably describes the content of that document well. For example, in a document about 'apple', the word 'apple' would appear many times.
    *   **Calculation**: There are several ways to calculate TF, but a common method is the raw count of a term in a document, or the frequency normalized by the total number of words in the document to prevent bias towards longer documents. For example, `TF(t, d) = (Number of times term t appears in document d) / (Total number of terms in document d)`.

2.  **Inverse Document Frequency (IDF)**
    *   **Concept**: Inverse Document Frequency measures how important a term is across the entire corpus. It down-weights terms that appear very frequently across many documents and up-weights terms that appear rarely across the corpus.
    *   **Intuition**: Common words like "the", "is", "a" (stopwords) appear in almost every document and thus carry little unique information about a document's specific topic. Conversely, a word that appears in only a few documents is likely to be very distinctive and important for those documents. IDF helps to filter out these common, less informative words.
    *   **Calculation**: `IDF(t, D) = log_e(Total number of documents D / Number of documents containing term t)`. A small constant (usually 1) is often added to the denominator to prevent division by zero if a term does not appear in any document in the corpus.

### How TF-IDF Combines These Metrics

TF-IDF is calculated by multiplying TF and IDF:

`TF-IDF(t, d, D) = TF(t, d) * IDF(t, D)`

*   `t`: The term (word)
*   `d`: The specific document
*   `D`: The corpus (collection of documents)

**The intuition behind multiplying TF by IDF is to identify relevant and unique terms:**

*   If a word appears frequently in a *specific document* (high TF) but rarely across the *entire corpus* (high IDF), its TF-IDF score will be high. This indicates that the word is highly characteristic of that document and helps differentiate it from other documents.
*   If a word appears frequently in a *specific document* (high TF) but also frequently across the *entire corpus* (low IDF), its TF-IDF score will be moderate or low. This signifies that the word is common and not very distinctive for that document.
*   If a word appears rarely in a *specific document* (low TF), its TF-IDF score will be low, regardless of its IDF. This means the word is not central to that document's content.

## Explain Sigmoid Function Intuition

### Subtask:
Explain the intuition and role of the sigmoid activation function in Logistic Regression.

#### Explanation of Sigmoid Function Intuition and Role in Logistic Regression

The **Sigmoid function**, also known as the **logistic function**, is a crucial component in Logistic Regression. Its primary role is to map any real-valued number into a value between 0 and 1, making it ideal for converting a linear combination of inputs into a probability.

1.  **Intuition and Mathematical Formula**

    The intuition behind the sigmoid function is to provide a smooth, S-shaped curve that can represent a probability. It is especially useful when we want to model the probability of a binary outcome (e.g., 0 or 1, true or false, positive or negative).

    The mathematical formula for the sigmoid function, often denoted as $\sigma(z)$, is:

    $$\sigma(z) = \frac{1}{1 + e^{-z}}$$

    Where:
    *   $z$ is the input to the function, which can be any real number.
    *   $e$ is Euler's number (approximately 2.71828).

2.  **Output Range and Relationship to Probabilities**

    The output range of the sigmoid function is always between **0 and 1** (exclusive). That is, $0 < \sigma(z) < 1$.

    *   As $z$ approaches positive infinity ($z \to \infty$), $e^{-z}$ approaches 0, so $\sigma(z)$ approaches 1.
    *   As $z$ approaches negative infinity ($z \to -\infty$), $e^{-z}$ approaches positive infinity, so $\sigma(z)$ approaches 0.
    *   When $z = 0$, $e^{-0} = 1$, so $\sigma(0) = \frac{1}{1 + 1} = 0.5$.

    This output range is perfectly suited for representing probabilities. In Logistic Regression, the output of the sigmoid function is interpreted as the probability that a given input belongs to a certain class (e.g., the probability of being 'positive').

3.  **Role in Logistic Regression**

    In Logistic Regression, the model first calculates a linear combination of the input features and their corresponding weights. This linear combination can be represented as $z = w_0 + w_1x_1 + w_2x_2 + ... + w_nx_n$, or more compactly as $z = \mathbf{w}^T\mathbf{x}$, where $\mathbf{w}$ is the vector of weights and $\mathbf{x}$ is the vector of input features.

    This $z$ value can be any real number, positive or negative. However, to transform this linear score into a probability for binary classification, we pass it through the sigmoid function.

    The sigmoid function then converts this $z$ value into a probability $P(Y=1|X)$, which represents the likelihood of the instance belonging to the positive class (e.g., label = 'positive'):

    $$P(Y=1|X) = \sigma(\mathbf{w}^T\mathbf{x}) = \frac{1}{1 + e^{-(\mathbf{w}^T\mathbf{x})}}$$

    If this probability is greater than a certain threshold (commonly 0.5), the instance is classified as belonging to the positive class; otherwise, it's classified as belonging to the negative class. This transformation allows Logistic Regression to perform binary classification by modeling the probability of an event occurring.
